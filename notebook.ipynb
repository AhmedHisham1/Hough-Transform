```python
import numpy as np
import matplotlib.image as mpimg
import cv2
from matplotlib import pyplot as plt

%matplotlib inline
```

```python
black_img = np.zeros((300,300))

# Drawing lines on this black image
x = np.linspace(0, 299, 300).astype(np.int32)
m, b = 0.5, 120
y = m*x + b
y = y.astype(np.int32)[y>=50]

img = black_img
img[y,x] = 1

y = 0.5*x + 0
img[y.astype(np.int32), x] = 1
```

```python
plt.figure(figsize=(6,6))
plt.imshow(img, cmap='gray')
```

```python
from utils import hough_transform
A = hough_transform(img, theta_step=0.004)
plt.figure(figsize=(10,10))
plt.imshow(A)
```
